# Fitten in de fitneeeessss

```python
def least_squared(theta, f, x, y, sigma_y):
    elements = (y - f(x, theta))**2 / sigma_y**2    
    return np.sum(elements)
```

```python
def model(W, theta):
    X = theta[0]
    b = theta[1]
    
    return X/np.sqrt(W**4 + 2*(b**2 - 1) + 1)
```

```python
def find_parameters(f, x, y, sigma_y, initial_guess):
    return opt.minimize(least_squared, initial_guess,  args = (f, x, y, sigma_y))
```

```python
oplossing = find_parameters(model, f, a, da, (10,10))
```
